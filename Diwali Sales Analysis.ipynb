```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Diwali Sales Data.csv' , encoding="unicode_escape")
df
```

```python
df.info()
```

```python
# dropping the empty columns

df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.info()
```

```python
df["Amount"].isna().sum()
```

```python
# Dropping the null values rows.

df.dropna(inplace=True)
df
```

```python
df.index = range(0,11239)
```

```python
# Changing the datatype of Amount column from float to int

df["Amount"] = df["Amount"].astype(int)
df.dtypes
```

## Gender

```python
# Visualising the number of males and females.

plt.figure(figsize=(5,5))
countplot = sns.countplot(x="Gender",data=df)

for bars in countplot.containers:           # To add label to the graph
    countplot.bar_label(bars)
```

```python
# Finding out the amount spent by males and females.

spendings_gender = df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
spendings_gender
```

```python
# Visualising Amount spent by Males and Females.

plt.figure(figsize=(5,5))
sns.barplot(x="Gender",y="Amount",data=spendings_gender)
```

#### As we can see, Number of Females is much higher as well as the amount spent by females is also more than men

## Age

```python
# Finding out the distribution of Buyers by age and gender

plt.figure(figsize=(10,8))
age_plot = sns.countplot(data=df,x="Age Group" , hue="Gender")

for bars in age_plot.containers:
    age_plot.bar_label(bars)
```

#### As we can see, Ages 26-35 comprise majority of the buyers. 

```python
# Finding the amount spent by age.

plt.figure(figsize=(10,5))
sales_age = df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

sns.barplot(x="Age Group", y="Amount", data=sales_age)
```

####  From the above graphs,  we can see that most buyers are in the age group 26-35 and most of them are females.

## State

```python
# finding out number of orders state wise

orders_state = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)

plt.figure(figsize=(20,10))
sns.barplot(data=orders_state,x="State",y="Orders")
```

```python
amount_state = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

plt.figure(figsize=(20,10))
sns.barplot(data=amount_state,x="State",y="Amount")
```

#### From the above graphs we can conclude that uttar pradesh has the most numbers of orders as well as most amount spent.

## Marital Status

```python
# Finding out the count of married and unmarried people.

plt.figure(figsize=(5,5))
marital_countplot = sns.countplot(data=df,x="Marital_Status")

for bars in marital_countplot.containers:
    marital_countplot.bar_label(bars)
```

```python
# Finding the amount spent by females and males as per their marital status.

plt.figure(figsize=(10,5))
sales_gender = df.groupby(["Marital_Status","Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

sns.barplot(data=sales_gender,x="Marital_Status", y="Amount", hue="Gender")
```

#### From the above Graphs, we can conclude that most of the buyers are married and most amount is spent by Women.

## Occupation 

```python
# Finding out the number of buyers as per their occupation.

plt.figure(figsize=(20,10))
occupation_countplot = sns.countplot(data=df,x="Occupation")

for bars in occupation_countplot.containers:
    occupation_countplot.bar_label(bars)
```

```python
# Finding out the amount spent by each occupation sector.

plt.figure(figsize=(20,10))
sales_occupation = df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

sns.barplot(data=sales_occupation,x="Occupation",y="Amount")
```

#### As per the graphs, IT sector is the highest spender and agriculture sector is the lowest spender.

## Product category

```python
# Finding out the number of orders as per the product category

plt.figure(figsize=(25,10))
product_countplot = sns.countplot(data=df,x="Product_Category")

for bars in product_countplot.containers:
    product_countplot.bar_label(bars)
```

```python
# finding the amount spent on each of the product category.

plt.figure(figsize=(25,10))
sales_product = df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

sns.barplot(data=sales_product,x="Product_Category",y="Amount")
```

#### From the above graphs we can conclude that the most sold product categories are clothing, food and electronics
#### However the most amount spent on categories is in the order of food, clothing and electronics

## Product ID

```python
# Finding out the top 10 selling products by product id

plt.figure(figsize=(20,7))
orders_products = df.groupby(["Product_ID"], as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)

sns.barplot(data=orders_products,x="Product_ID", y="Orders")
```
